```python
import os

import pandas as pd

json_files_directory = os.path.join("outputs", "offres")
json_name = "_offres_concatenated_13639_offres__2025-03-05--22h09.json"
```

```python
pd.options.display.max_rows = 3000
pd.options.display.max_columns = 100
```

```python
df = pd.read_json(
    os.path.join(json_files_directory, json_name),
    dtype=False,
)
```

```python
df.head(2)
```

```python
df.shape
```

```python
lieuTravail_normalized = pd.json_normalize(df["lieuTravail"])

df = df.join(lieuTravail_normalized)

df_lieu = df[["id", "intitule"] + list(lieuTravail_normalized.columns)]
```

```python
df_lieu
```

#### <u> Cas 01 : quand libelle = ("FRANCE"|"France"|"France entière") (dans ce cas code postal = code commune = NAN), et latitude = longitude = NAN </u>

```python
# df_lieu[~df_lieu.libelle.str.match(r"^\d{2}\s-\s")].value_counts("libelle")


```

```python
cas_01 = df_lieu[df_lieu.libelle.isin(["FRANCE", "France", "France entière"]) \
                & df_lieu.latitude.isna() \
                & df_lieu.longitude.isna() \
                ]  # fmt:off
```

```python
cas_01
```

```python
# On écrit le numéro du cas dans "lieu_cas"
df_lieu.loc[df_lieu.id.isin(cas_01.id), "lieu_cas"] = "cas_01"
```

```python
df_lieu[df_lieu.lieu_cas == "cas_01"]
```

```python
df_lieu.value_counts("lieu_cas")
```

## <u> Cas 02 : quand libelle = ("FRANCE"|"France"|"France entière") (dans ce cas code postal = code commune = NAN), et latitude/longitude sont renseignés </u>

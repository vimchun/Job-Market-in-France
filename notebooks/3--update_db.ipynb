```python
import os

import pandas as pd

generated_json_files_directory = os.path.join("..", "api_extract__transform", "outputs", "offres", "1--generated_json_file")

json_filename_1 = "2025-04-03--18h36__extraction_occurence_12.json"
json_filename_2 = "2025-04-03--23h40__extraction_occurence_13.json"

df1 = pd.read_json(os.path.join(generated_json_files_directory, json_filename_1), dtype=False)  # représente "json_n-1"
df2 = pd.read_json(os.path.join(generated_json_files_directory, json_filename_2), dtype=False)  # représente "json_n"
```

```python
# pd.options.display.max_rows = 3000
pd.options.display.max_columns = 100
```

```python
display(
    df1.shape,
    df2.shape,
    df1.head(1),
    df2.head(1),
)
```

```python
intersection_ids = pd.merge(df1, df2, on="id")["id"].tolist()

df1_minus_intersection = df1[~df1.id.isin(intersection_ids)]

df_concat = pd.concat([df1_minus_intersection, df2])

display(
    df1_minus_intersection.shape,
    df2.shape,
    df_concat,
)
```

# <u> Exemple simple </u>

```python
df1 = pd.DataFrame({"id": [1, 2, 3, 4], "valeur": ["df1_A", "df1_B", "(df__C)", "(df__D)"]})
df2 = pd.DataFrame({"id": [3, 4, 5, 6], "valeur": ["(df__W)", "(df__X)", "df2_Y", "df2_Z"]})

intersection_ids = pd.merge(df1, df2, on="id")["id"].tolist()  ##==> [3, 4]

partie_1 = df1[~df1.id.isin(intersection_ids)]
partie_2 = df2

display(
    partie_1,
    partie_2,
    pd.concat([partie_1, partie_2]),
)
```

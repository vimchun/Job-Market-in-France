## <u> Cas_2 : "code_insee = NAN" (dans ce cas "code_postal = NAN"), mais coordonnées GPS renseignées </u>

### <u> Récupération code_postal avec geopy </u>

```python
df_lieu[~df_lieu.lieu_cas.isin(["cas_1"])  # != cas_1
      & ~df_lieu.latitude.isna()
        ]  # fmt: off
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_geopy_app")
```

```python
latitude_min = 42.3328  # Sud
latitude_max = 51.0892  # Nord
longitude_min = -4.7956  # Ouest
longitude_max = 8.2306  # Est


def check_and_swap(row):
    """
    Fonction pour inverser la latitude et la longitude si la latitude est hors de la fourchette
    """

    if row["latitude"] < latitude_min or row["latitude"] > latitude_max:
        row["latitude"], row["longitude"] = row["longitude"], row["latitude"]

    if (latitude_min <= row["latitude"] <= latitude_max) and (longitude_min <= row["longitude"] <= longitude_max):
        # on assigne "cas_2" à "lieu_cas" si les coordonnées sont dans la France métropolitaine
        row["lieu_cas"] = "cas_2"
    else:
        row["lieu_cas"] = "cas_2_coords_gps_hors_FR"

    try:
        """
        Ce morceau de code est long car la lib fait des requêtes...
        Pour info, 5 minutes pour 191 offres
        La lib geopy permet de récupérer le code postal (pas le code commune).
        Donc on récupère le code postal ici, puis on récupèrera le code commune ensuite pour avoir la ville, le département et la région.
        """
        location = geolocator.reverse((row["latitude"], row["longitude"]), language="fr", exactly_one=True)
        if location:
            address = location.raw.get("address", {})
            zipcode = address.get("postcode", "Inconnu")
            row["code_postal"] = zipcode
        else:
            row["code_postal"] = "Inconnu"
    except Exception as e:
        row["code_postal"] = "Erreur geocodage"
        print(f"Erreur pour les coordonnées ({row['latitude']}, {row['longitude']}): {e}")

    return row
```

```python
# /!\ temps d'exécution de plusieurs minutes
cas_2 = df_lieu[~df_lieu.lieu_cas.isin(["cas_1"])  # != cas_1
               & ~df_lieu.latitude.isna()
                  ].apply(check_and_swap, axis=1)  # fmt: off
```

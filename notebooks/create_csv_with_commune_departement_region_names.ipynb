```python
import os

import pandas as pd

files_directory = os.path.join("..", "additional_files", "archives", "lien_2")
file_commune = "v_commune_2024.csv"
file_departement = "v_departement_2024.csv"
file_region = "v_region_2024.csv"
```

## <u> df_commune </u>

```python
df_commune = pd.read_csv(
    os.path.join(files_directory, file_commune),
    usecols=[
        "COM",
        "REG",
        "DEP",
        "LIBELLE",
    ],
)
```

```python
df_commune.rename(
    {
        "COM": "code_commune",
        "REG": "code_region",
        "DEP": "code_departement",
        "LIBELLE": "nom_commune",
    },
    axis=1,
    inplace=True,
)
```

```python
df_commune
```

## <u> df_departement </u>

```python
df_departement = pd.read_csv(
    os.path.join(files_directory, file_departement),
    usecols=[
        "DEP",
        "LIBELLE",
    ],
)
```

```python
df_departement.rename(
    {
        "DEP": "code_departement",
        "LIBELLE": "nom_departement",
    },
    axis=1,
    inplace=True,
)
```

```python
df_departement
```

## <u> df_region </u>

```python
df_region = pd.read_csv(
    os.path.join(files_directory, file_region),
    usecols=[
        "REG",
        "LIBELLE",
    ],
)
```

```python
df_region.rename(
    {
        "REG": "code_region",
        "LIBELLE": "nom_region",
    },
    axis=1,
    inplace=True,
)
```

```python
df_region
```

## <u> merging </u>

```python
df_commune.head(2)
```

```python
df_departement.head(2)
```

```python
df_region.head(2)
```

```python
df = df_commune.merge(df_departement, on="code_departement").merge(df_region, on="code_region")
```

```python
df = df[["code_commune", "nom_commune", "nom_departement", "nom_region"]]
```

```python
df.to_csv(os.path.join(files_directory, "commune_departement_region_names.csv"))
```

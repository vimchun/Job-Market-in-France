```python
import os

import pandas as pd

json_files_directory = os.path.join("outputs", "_archives", "2025-03-02--exemples-jsons-et-json-concatenated")
json_name = "_offres_concatenated_13639_offres__2025-03-05--22h09.json"
```

```python
lieuTravail_normalized = pd.json_normalize(df["lieuTravail"])

df = df.join(lieuTravail_normalized)

df_lieu = df[["id", "intitule"] + list(lieuTravail_normalized.columns)]
```

## <u> Cas_2 : "code_commune = NAN" (dans ce cas "code_postal = NAN"), mais coordonnées GPS renseignées </u>

```python
df_lieu[~df_lieu.lieu_cas.isin(["cas_1"])  # != cas_1
      & ~df_lieu.latitude.isna()
        ]  # fmt: off
```

```python
latitude_min = 42.3328  # Sud
latitude_max = 51.0892  # Nord
longitude_min = -4.7956  # Ouest
longitude_max = 8.2306  # Est


def check_and_swap(row):
    """
    Fonction pour inverser la latitude et la longitude si la latitude est hors de la fourchette
    """

    if row["latitude"] < latitude_min or row["latitude"] > latitude_max:
        row["latitude"], row["longitude"] = row["longitude"], row["latitude"]

    if (latitude_min <= row["latitude"] <= latitude_max) and (longitude_min <= row["longitude"] <= longitude_max):
        # on assigne "cas_2" à "lieu_cas" si les coordonnées sont dans la France métropolitaine
        row["lieu_cas"] = "cas_2"
    else:
        row["lieu_cas"] = "cas_2_coords_gps_hors_FR"

    return row
```

```python
cas_2 = df_lieu[~df_lieu.lieu_cas.isin(["cas_1"])  # != cas_1
               & ~df_lieu.latitude.isna()
                  ].apply(check_and_swap, axis=1)  # fmt: off
```

```python
cas_2.head()
```

```python
# On met à jour df_lieu avec les valeurs de latitude/longitude (car certaines offres ont vu ces valeurs s'inverser), et la valeur de lieu_cas
df_lieu.update(cas_2)
```
